```python
import xml.etree.ElementTree as ET
import pandas as pd
import os

# --- 1. Define Paths ---
base_path = '/content/drive/MyDrive/AI_Physio_Coach/dataset/'
annot_path = os.path.join(base_path, 'annotatorA')

# --- 2. Create an empty list to hold our data ---
# Each item in the list will be a dictionary for one repetition
v2_data_list = []

# --- 3. Loop through all files in the annotation directory ---
print(f"Starting to process files in: {annot_path}")
file_list = os.listdir(annot_path)
processed_files = 0
skipped_files = 0

for filename in file_list:
    # Only process the .anvil (XML) files
    if not filename.endswith('.anvil'):
        continue

    # Create the full path to the file
    full_file_path = os.path.join(annot_path, filename)

    try:
        # --- 4. Parse the XML ---
        tree = ET.parse(full_file_path)
        root = tree.getroot()

        # --- 5. Find the 'Global evaluation' track ---
        # We use an XPath query to find all 'el' elements within this specific track
        annotations = root.findall(".//track[@name='Global evaluation']/el")

        if not annotations:
            # If no 'Global evaluation' track, try 'Specific Errors' as a backup
            annotations = root.findall(".//track[@name='Specific Errors']/el")

        # --- 6. Extract Data for each repetition (<el>) ---
        for el in annotations:
            label_node = el.find("attribute[@name='evaluation']")

            # Check if we have all the data we need
            if label_node is not None and 'start' in el.attrib and 'end' in el.attrib:
                label = label_node.text
                start_time = float(el.attrib['start'])
                end_time = float(el.attrib['end'])

                # The skeleton filename matches the anvil filename, but with a different extension
                # (e.g., .json for OpenPose). We'll store the base name.
                base_filename = filename.replace('.anvil', '')

                # --- 7. Append to our master list ---
                v2_data_list.append({
                    'filename': base_filename,
                    'start_time': start_time,
                    'end_time': end_time,
                    'label': label
                })

        processed_files += 1

    except Exception as e:
        print(f"Error processing file {filename}: {e}")
        skipped_files += 1

print("\n--- Processing Complete ---")
print(f"Successfully processed {processed_files} files.")
print(f"Skipped {skipped_files} files due to errors.")

# --- 8. Convert the list of dictionaries into a Pandas DataFrame ---
if v2_data_list:
    df_v2_labels = pd.DataFrame(v2_data_list)

    print("\n--- Dataset Created Successfully ---")
    print(f"Total repetitions (labels) found: {len(df_v2_labels)}")

    print("\n--- Sample of the data (first 5 rows): ---")
    print(df_v2_labels.head())

    print("\n--- Unique Error Labels Found ---")
    print(df_v2_labels['label'].value_counts())
else:
    print("No data was loaded. Please check file paths and XML structure.")
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
import os

# --- 1. Define Paths ---
base_path = '/content/drive/MyDrive/AI_Physio_Coach/dataset/'
annot_path_B = os.path.join(base_path, 'annotatorB')

# --- 2. Create an empty list to hold our data ---
v2_data_list_B = []

# --- 3. Loop through all files in the annotatorB directory ---
print(f"Starting to process files in: {annot_path_B}")
file_list_B = os.listdir(annot_path_B)
processed_files_B = 0
skipped_files_B = 0

for filename in file_list_B:
    if not filename.endswith('.anvil'):
        continue

    full_file_path = os.path.join(annot_path_B, filename)
    base_filename = filename.replace('.anvil', '')

    try:
        tree = ET.parse(full_file_path)
        root = tree.getroot()

        global_eval_reps = root.findall(".//track[@name='Global evaluation']/el")

        if not global_eval_reps:
            skipped_files_B += 1
            continue

        for rep_el in global_eval_reps:
            main_label_node = rep_el.find("attribute[@name='evaluation']")
            if main_label_node is None:
                continue

            main_label = main_label_node.text
            start_time = float(rep_el.attrib['start'])
            end_time = float(rep_el.attrib['end'])

            final_label = main_label

            if main_label == 'Incorrect':
                error_track = root.find(".//track[@name='Global error']")
                if error_track is not None:
                    error_el = error_track.find('el')
                    if error_el is not None:
                        specific_error_node = error_el.find("attribute[@name='type']")
                        if specific_error_node is not None:
                            final_label = specific_error_node.text

            v2_data_list_B.append({
                'filename': base_filename,
                'start_time': start_time,
                'end_time': end_time,
                'label': final_label
            })

        processed_files_B += 1

    except Exception as e:
        print(f"Error processing file {filename}: {e}")
        skipped_files_B += 1

print("\n--- Processing Complete (Annotator B) ---")
print(f"Successfully processed {processed_files_B} files.")
print(f"Skipped {skipped_files_B} files.")

# --- 4. Convert the list of dictionaries into a Pandas DataFrame ---
if v2_data_list_B:
    df_v2_labels_B = pd.DataFrame(v2_data_list_B)

    print("\n--- Annotator B Dataset Created Successfully ---")
    print(f"Total repetitions (labels) found: {len(df_v2_labels_B)}")

    print("\n--- ⭐️ NEW Unique Error Labels Found (Annotator B) ---")
    print(df_v2_labels_B['label'].value_counts())
else:
    print("No data was loaded from Annotator B.")
```

```python
import pandas as pd
import os

# --- 1. Load the Combined Label DataFrame ---
base_path = '/content/drive/MyDrive/AI_Physio_Coach/dataset/'
label_file_path = os.path.join(base_path, 'df_v2_master_labels.csv')
df_v2_master_labels = pd.read_csv(label_file_path)

print(f"Loaded master label list with {len(df_v2_master_labels)} total samples.")

# --- 2. Clean the Labels ---
# We will create a list of the "junk" labels we want to remove.
# For now, this is just 'Incorrect', but we can add 'Other' or 'Motionless' if we decide to.
labels_to_remove = ['Incorrect']

# Create the new, clean DataFrame by selecting only the rows
# where the label is NOT in our 'labels_to_remove' list.
df_v2_labels_clean = df_v2_master_labels[
    ~df_v2_master_labels['label'].isin(labels_to_remove)
].copy()

# --- 3. Report the Results ---
print(f"Removed {len(df_v2_master_labels) - len(df_v2_labels_clean)} samples with generic labels.")
print(f"Our final V2 dataset will have {len(df_v2_labels_clean)} high-quality labeled samples.")

print("\n--- ⭐️ FINAL V2 Unique Labels (Cleaned) ---")
print(df_v2_labels_clean['label'].value_counts())

# --- 4. Save the Final Clean Label List ---
# We save this to a new file, which will be our instruction manual for the next step.
clean_save_path = os.path.join(base_path, 'df_v2_labels_CLEAN.csv')
df_v2_labels_clean.to_csv(clean_save_path, index=False)

print(f"\nFinal, clean V2 label list saved to: {clean_save_path}")
```
